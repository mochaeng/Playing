# Visualizing the data

```python
def display_hist(ddf, column_name, bins):
    plt.figure(figsize=(8, 6))
    n, bins, patches = plt.hist(ddf[column_name].compute(), bins=bins, color='blue', alpha=0.7)

    for bin_val, count in zip(bins, n):
        if count == 0:
            continue
        plt.text(bin_val + (bins[1] - bins[0]) / 2, count, str(int(count)),
                ha='center', va='bottom', color='black', fontsize=8)

    plt.title(f'Histogram of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()
```

### Label Class

```python
display_hist(ddf, 'Label', 3)
```

### Attack class

```python
display_hist(ddf, 'Attack', 20)
```
