# Removing Columns

Removing desiring columns and choosing a random sample of the original dataset

```python
ddf = ddf.drop(columns=['IPV4_SRC_ADDR', 'IPV4_DST_ADDR'])
ddf = ddf.sample(frac=0.147575372055222, random_state=69)
```

# Saving the dataset

Saving the new dataset into parquet format to be more efficient

```python
ddf.repartition(npartitions=1).to_parquet('./data/csv-to-parquet/X_NF-ToN-IoT-V2.parquet.gzip', compression='gzip')
```
